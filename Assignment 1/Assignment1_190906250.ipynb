## Q3. (4 points)

Design a mapping function that will convert the status column value (i.e. developing, developed ) to a numerical value (0, 1).


```python
data = pd.read_csv(DATA_FILE_PATH)
statusMap = {"Developing": 0, "Developed": 1}

data["Status"] = data["Status"].map(statusMap)
data
```

## Q4. (5 points)

What are the minimum, average and maximum GDP in the developing country. 
List the country name.

```python
developingCountries = data.loc[data["Status"] == 0]
gdp = developingCountries["GDP"]

maxGdp = developingCountries.loc[gdp.idxmax()]
minGdp = developingCountries.loc[gdp.idxmin()]
meanGdp = gdp.mean()

print(f"Maximum GDP:\n{maxGdp}\n\n")
print(f"Minimum GDP:\n{minGdp}\n\n")
print(f"Mean GDP:\n{meanGdp}")
```
